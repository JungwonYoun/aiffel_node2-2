## 택시의 주행 시간과 주행 거리, 요금 등의 상관 관계 알아보기

```python
# 주행 시간 컬럼 만들기
trip_df['tpep_pickup_datetime'] = pd.to_datetime(trip_df['tpep_pickup_datetime'])
trip_df['tpep_dropoff_datetime'] = pd.to_datetime(trip_df['tpep_dropoff_datetime'])
trip_df['driving_time'] = trip_df['tpep_dropoff_datetime'] - trip_df['tpep_pickup_datetime']
trip_df['driving_time'] = (trip_df['driving_time'].dt.seconds.div(60).astype(int) 
             + trip_df['driving_time'].dt.days.multiply(1440).astype(int) )
trip_df
```

```python
# 주행시간과 요금 컬럼으로 그래프 그리기
# 주행 거리와 요금컬럼으로 그래프 그리기
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('fare (s)')
ax1.set_ylabel('time', color=color)
ax1.scatter(trip_df['fare_amount'], trip_df['driving_time'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('distance', color=color)  # we already handled the x-label with ax1
ax2.scatter(trip_df['fare_amount'], trip_df['trip_distance'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
